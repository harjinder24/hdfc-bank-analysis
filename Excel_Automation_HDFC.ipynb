```python
# columns to pass in the cleaned excel workbook
columns = ['bank_name', 'bank_type', 'credit_cards', 'debit_cards', 'cc_transactions', 'cc_inr_value', 
           'dc_transactions', 'dc_inr_value', 'month']

# Creating a new blank excel workbook with desired column names
main_wb = openpyxl.Workbook()
bnk_sheet = main_wb.active

for col_index, value in enumerate(columns, start=1):
    cell = bnk_sheet.cell(row=1, column=col_index)
    cell.value = value
    
main_wb.save('/HDFC Analysis/Dataset/test/cleaned.xlsx')
```

```python
# loading banks data workbook
atm_wb_path = '/HDFC Analysis/Dataset/test/bank_sep_2022.xlsx'
atm_wb = openpyxl.load_workbook(atm_wb_path)

# activating the workbook
sheet_atm = atm_wb.active
```

```python
# Unmerging columns

merged_ranges = sheet_atm.merged_cells.ranges.copy()

for merged_range in merged_ranges:
    sheet_atm.unmerge_cells(merged_range.coord)
    
atm_wb.save(atm_wb_path)
```

```python
# Deleting unnecessary columns
columns_to_del = [4,5,6,7,8,9]

# reversing columns to avoid column indexing change
columns_to_del.sort(reverse=True)

for index in columns_to_del:
    sheet_atm.delete_cols(index)
    
atm_wb.save(atm_wb_path)
```

```python
# Deleting unnecessary rows

# fetching row number for payments bank and delete everything after payments bank
r_index = None
c_index = None

for row_ind, row in enumerate(sheet_atm.iter_rows(), start=1):
    for col_ind, cell in enumerate(row, start=1):
        if cell.value == 'Payment Banks':
            r_index = row_ind
            c_index = col_ind
            break
        if r_index is not None:
            break
            
# deleting additional rows
sheet_atm.delete_rows(r_index, sheet_atm.max_row-r_index)
atm_wb.save(atm_wb_path)
```

```python
# # Fetching bank names and bank type -- no need to run this here for cleaned sheet
main_wb_path = '/HDFC Analysis/Dataset/test/cleaned.xlsx'

bnk_type = ['Public Sector Banks', 'Private Sector Banks', 'Foreign Banks']
mr = sheet_atm.max_row
bt = None
bn = None
for ind in range(1, mr): 
    print("Fetching values...")
    br = sheet_atm.cell(row=ind, column=2).value  # Reading values from the second column
    cr = sheet_atm.cell(row=ind, column=3).value  # Reading values from the third column
    
    # checking bank type and bank names and extracting relevant value
    if br in bnk_type:
        bt_list = br.split(" ")
        bt = bt_list[0] # fetching bank type
        print(bt)
        continue
    elif str(br).isdigit():
        bn = cr # fetching bank name
        
    print(f"{bn} : {bt}")
        
    # appending bank name and bank type in new sheet
    print("Adding values to cleaned sheet")
    wb = openpyxl.load_workbook(main_wb_path)
    sheet = wb.active
    wmr = sheet.max_row
    bn_cell = sheet.cell(row=wmr+1, column=1)
    bn_cell.value = bn
    
    bt_cell = sheet.cell(row=wmr+1, column=2)
    bt_cell.value = bt
    print("Successfully added")
    
    wb.save(main_wb_path)
```
